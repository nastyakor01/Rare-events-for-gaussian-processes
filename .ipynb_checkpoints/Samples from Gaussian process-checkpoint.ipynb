```python

def get_cov(f, points1, points2):
    """Return covariance matrix for the sets of points given according to covariance function f"""
    size1 = points1.shape[0]
    size2 = points2.shape[0]
    cov = np.empty((size1, size2)) 
    for i in range(size1):
        for j in range(size2):
            cov[i, j] = f(points1[i], points2[j])
    return cov  
```

```python
# examples of covariance functions 

def K_Linear(s,t):
    return abs(s-t)

def Sinc(s, t):
    return np.sinc(s-t)

def K_Wiener(s, t):
    return min(s, t)

def K_RBF(s, t):
    return np.exp(-0.5 * abs(s-t)**2)

def K_brownian_bridge(s, t):
    return min(s, t) - s*t

def exponential(s, t):
    return np.exp(-abs(s-t))

def wave(s,t):
    return np.cos(s-t)

a = np.linspace(2, 5, 4)
b = np.linspace(4, 7, 5)
print(a)
print(b)
get_cov(wave, a, b)

def ift_triangle(s, t):
    if abs(s-t) <= 10e-3:
        return 1 - (s-t)**2/24
    else:
        return 2*(1 - np.cos(s-t))/(s-t)**2/np.sqrt(2*np.pi)
    
def ift_exponential(s, t):
    return np.sqrt(2/np.pi)/((s-t)**2 + 1)
```

```python
# examples of mean functions 
def square(s):
    return s*s

def linear(s):
    return s

def zero(s):
    return 0
```

```python
def simulate_GP(mean_function, cov_function, nb_of_samples, number_of_functions, left_border, right_border, ls = True):
    points = np.linspace(left_border, right_border, nb_of_samples)
    Σ = get_cov(cov_function, points, points)
    mean_vals = [mean_function(point) for point in points]
    ys = np.random.multivariate_normal(
                                       mean_vals,
                                       cov=Σ,
                                       size=number_of_functions)
    fig = plt.figure(figsize=(12, 6), dpi=80)
    
    for i in range(number_of_functions):
        if ls:
            plt.plot(points, ys[i])
            plt.plot(points, np.zeros_like(points), color = 'green') 
        #else:
        #    plt.plot(points, ys[i], ls = 'None')
        #plt.scatter(points, ys[i])      
    return ys, fig
```

```python
def prime(f,x):
    e = 10e-8
    return (f(x+e) - f(x-e) )/(2*e)
   
def prime_prime(f, x):
    e = 10e-4
    return ( f(x+e) - 2*f(x) + f(x-e) )/e/e
```

```python
def cube(x):
    return x*x*x

prime_prime(cube, 3)
```

```python
def E(cov_function):
    r = lambda x: cov_function(0, x) # covariance function 
    
    deriv = -prime_prime(r, 0) # minus second derivative
    alfa = np.sqrt(deriv/(r(0)+10e-8))
    E_zeros = 1/np.pi * alfa # count average amount of zeroes using Kac-Rice formula
        
    return E_zeros    
```

```python
# comparison function
def cmp(a, b): 
    if a > b:
        return 1
    elif a < b:
        return -1
    else:
        return 0


def intersections(mean_function, cov_function, nb_of_samples, number_of_functions, left_border, right_border, ls = True, y = True):
    zeros = np.zeros(number_of_functions) # будущее кол-во нулей по всем функциям 
    Y, _ = simulate_GP(mean_function, cov_function, nb_of_samples, number_of_functions, left_border, right_border, ls)
    size = Y.shape[1]
    
    # считаю кол-во нулей
    for j in range(number_of_functions): # если 2 соседних значения разных знаков, то между ними ноль
        pairs = [(Y[j][i], Y[j][i+1]) for i in range(size - 1)]  # буду проверять каждую пару значений 
        for (a,b) in pairs:
            if cmp(a, 0) != cmp(b, 0):
                zeros[j] += 1
    
    E_zeros = E(cov_function) * (right_border - left_border) # считаю среднее число нулей по формуле
    
    # и округляю до ближайшего целого 
    if np.ceil(E_zeros) - E_zeros > E_zeros - np.floor(E_zeros): 
        E_zeros = np.floor(E_zeros)
    else:
        E_zeros = np.ceil(E_zeros)
    
    if ls:
        plt.figure(figsize = (10, 5), dpi=80)
        plt.hist(zeros, edgecolor='white')
        print('E_zeros:',E_zeros)
    if y:
        return zeros, E_zeros, Y
    else:
        return zeros, E_zeros
```

# Solution 1.5

```python
# Gaussian process posterior
def Gaussian_Posterior(X1, y1, X2, kernel_func):
    """
    Calculate the posterior mean and covariance matrix for y2
    based on the corresponding input X2, the observations (y1, X1), 
    and the prior kernel function.
    """
    # Kernel of the observations
    N = X1.shape[0]
    add = np.eye(N)
    Σ11 = get_cov(kernel_func, X1, X1) + add*10e-8
    # Kernel of observations vs to-predict
    Σ12 = get_cov(kernel_func, X1, X2)
    # Solve
    solved = scipy.linalg.solve(Σ11, Σ12, assume_a='pos').T
    # Compute posterior mean
    μ2 = solved @ y1
    # Compute the posterior covariance
    Σ22 = get_cov(kernel_func, X2, X2)
    Σ2 = Σ22 - (solved @ Σ12)
    return μ2, Σ2  # mean, covariance
```

```python
def random_posterior(mean_function, cov_function, nb_of_samples, 
                     number_of_functions,
                     left_border,
                     right_border,
                     iterations,
                     ls = True,
                    ):
    """Create number_of_samples gaussian values on interval
       [left_border, right_border]
       according to prior distribution of the previous
       values, generated by means of Simulate_GP. 
    """
    
    """Здесь на первой итерации генерируется nb_of_samples значений обычным ковариационным методом
       (В той версии, которую я оставила, абсциссы делят временной интервал на равные части.
       Была еще версия, где они выбираются случайно равномерно, но там совсем плохо работает),
       потом на каждой следующей итерации случайно равномерно выбираются иксы, и в них генерятся игреки
       из нового условного распределения"""
    
    Graph = [] # будущий массив пар точек абсцисс и значений (x,y). С его помощью будем строить график 
    #X1 = []
    #for i in range(nb_of_samples):
    #    X1.append(np.random.uniform(left_border, right_border)) 
    X1 = np.linspace(left_border, right_border, nb_of_samples) # первая итерация. берем nb_of_samples иксов
    y1, fig = simulate_GP(mean_function, # генерим в них игреки с заданной ковариационной матрицей
                          cov_function,
                          nb_of_samples,
                          number_of_functions,
                          left_border,
                          right_border,
                          ls = False)
    y1 = np.squeeze(y1) # simulate_GP может строить много функций за один вызов, поэтому у игрека есть лишняя размерность.
    for k in range(nb_of_samples):
        Graph.append((X1[k], y1[k]))
    
    for iter in range(iterations):
        X2 = [] 
        for i in range(nb_of_samples): # каждую итерацию случайно равномерно выбираем иксы
            X2.append(np.random.uniform(left_border, right_border)) 
        μ2, Σ2 = Gaussian_Posterior(np.array(X1), y1, np.array(X2), cov_function) # пересчитываем распределение
        y2 = np.random.multivariate_normal(μ2, cov=Σ2) # из него берем игреки
        for k in range(nb_of_samples): 
            Graph.append((X2[k], y2[k]))
        X1 = X2
        y1 = y2
        
    Graph = sorted(Graph, key = lambda x: x[0])
    X = [g[0] for g in Graph]
    y = [g[1] for g in Graph]
    if ls:       
        plt.plot(X, y, figure = fig)
        plt.plot(X, np.zeros_like(X), color = 'green')     
        plt.show() 
    return y, fig     
```

Если итераций мало, а nb_of_samples большой, то все выглядит нормально. Иначе -- появляются сильные скачки.

```python
def Posterior3(mean_function, cov_function, nb_of_samples, number_of_functions, left_border, right_border, iterations, ls = True):
    """Create number_of_samples gaussian values on interval
       [left_border, right_border]
       according to prior distribution of the previous
       values, generated by means of Simulate_GP. 
    """
    
    len_interval = (right_border - left_border) / (iterations+1)
    delta = len_interval/(nb_of_samples - 1)

    final_y = np.zeros((number_of_functions, (iterations+1)*nb_of_samples))
    # create initial values using solution 1
    X1 = np.linspace(left_border, left_border+len_interval, nb_of_samples)
    y1, fig = simulate_GP(mean_function, 
                          cov_function,
                          nb_of_samples,
                          number_of_functions,
                          left_border,
                          left_border+len_interval,
                          ls = False)
    
    for func in range(number_of_functions): 
        y = np.zeros((iterations+1, nb_of_samples))
        X = np.zeros((iterations+1, nb_of_samples))
        y[0,:] = y1[func]
        X[0,:] = X1
        for it in range(1, iterations+1):
            X2 = np.linspace(left_border + len_interval*it, left_border + len_interval*(it+1), nb_of_samples)
            #compute posterior distribution 
            μ2, Σ2 = Gaussian_Posterior(X[it-1,:], y[it-1,:], X2, cov_function)
            #generate new samples
            y[it, :] = np.random.multivariate_normal(μ2, cov=Σ2)
            X[it, :] = X2
            
        X = np.reshape(X,((iterations+1)*nb_of_samples,))  
        final_y[func] = np.reshape(y,((iterations+1)*nb_of_samples,))
        if ls:       
            plt.plot(X, final_y[func], figure = fig)
            plt.plot(X, np.zeros_like(X), color = 'green')     
         
    if ls:
        plt.show() 
    return final_y, fig  
```

```python
def post_intersections(mean_function, cov_function, nb_of_samples, number_of_functions, left_border, right_border, iterations, ls = True, y = False):
    zeros = np.zeros(number_of_functions) # future amount of zeros for all functions  
    Y, _ = Posterior3(mean_function,
                      cov_function,
                      nb_of_samples,
                      number_of_functions,
                      left_border,
                      right_border,
                      iterations,
                      ls)
    size = Y.shape[1]
    
    # count amount of zeros
    for j in range(number_of_functions): # if two neighbouring values are of different signs, then there's a zero between them
        pairs = [(Y[j][i], Y[j][i+1]) for i in range(size - 1)]  # check every pair of neighbouring values 
        for (a,b) in pairs:
            if cmp(a, 0) != cmp(b, 0): # compare two values
                zeros[j] += 1
    
    E_zeros = E(cov_function) * (right_border - left_border) # count average amount of zeroes using Kac-Rice formula

    if ls: 
        plt.figure(figsize = (10, 5), dpi=80)
        plt.hist(zeros, edgecolor='white')
        print('E_zeros:',E_zeros)
    if y: 
        return zeros, E_zeros, Y
    else:
        return zeros, E_zeros
```

```python
# Хочу посмотреть, какую плотность порождают функции с андеркраудингом, и как она отличается от настоящей плотности процесса.
# В этой ячейке ядро -- гауссиан, поэтому плотность с ним совпадает. 
# Строю 4 графика:
# синий (under_density) -- плотность порождаемая процессами с антеркраудингом 
# зеленый (empirical_density) -- плотность порождаемая всеми процессами. То есть они с синим должны отличаться.
# желтый (true_density) -- реальная плотность, построенная как фурье от ядра. Должна быть очень похожа на зеленый.
# красный -- ядро. Должно быть точь-в-точь как желтый.


kernel = K_RBF # covariance functions
r = lambda x: kernel(0, x)

Eps = [0.01, 0.05]#, 0.08, 0.15] # epsilon
T = [50, 100]#, 5000, 10000] # time intervals 

for i, e in enumerate(Eps):
    for j, t in enumerate(T): 
        # Y -- массив из 100 функций, zeros -- массив нулей, E_zeros -- среднее число нулей
        zeros, E_zeros, Y = post_intersections(zero, kernel, 20, 100, -t/2, t/2, t//4, ls = False, y = True) 
        undercrowding = np.zeros(Y.shape[1])
        Q = np.fft.fft(undercrowding) # будущий under_density
        Q = np.array([np.real(q) for q in Q])
        D = np.fft.fft(undercrowding) # будущий empirical_density
        D = np.array([np.real(d) for d in D])
        
        size = Q.shape[0]
        N = np.linspace(-size//2, size//2, size+1)[:-1]
        count = 0
        
        for z in range(len(zeros)):
            D += np.array([np.abs(d)**2 for d in np.fft.fft(Y[z])])
            if E_zeros - zeros[z] < -e*t:
                Q0 =  np.fft.fft(Y[z])
                Q0 = np.array([np.abs(q)**2 for q in Q0])
                Q0 /= size
                Q += Q0
                count += 1
                
        Q /= count + 1e-30
        Q *= t/size/np.sqrt(2*np.pi)
        Q = np.hstack((Q[size//2:], Q[:size//2]))
        under_density = Q
        
        D /= len(zeros)
        D /= size
        D *= t/size/np.sqrt(2*np.pi)
        D = np.hstack((D[size//2:], D[:size//2]))
        empirical_density = D
        
        #axs[i, j].plot(N, under_density)
        #axs[i, j].plot(N, empirical_density)
        #axs[i, j].set_title(f'Eps = {e},T = {t}')
        X = np.linspace(-t/2, t/2, size+1)[:-1]
        k = [r(x) for x in X]                  # kernel time points
        true_density = np.fft.fft(k)
        true_density = [abs(np.real(x)) for x in true_density]
        true_density = np.hstack((true_density[size//2:], true_density[:size//2]))
        true_density *= t/size/np.sqrt(2*np.pi)
        
        
        
        delta = X[1]-X[0] # = A/N - time step
        F = 1./delta      # = N/A - sampling frequency
        w = np.linspace(-F/2, F/2, size+1)[:-1]   # FFT frequencies
        K = [r(2*np.pi*x) for x in w]          # Fourier transform of a kernel in 2pi*w points


        plt.plot(w, under_density, color = 'blue')
        plt.plot(w, empirical_density, color = 'green')
        plt.plot(w,  true_density, color = 'orange')
        plt.plot(w, K, color = 'red')
        
```

```python
def draw_over(kernel, Eps = [0.01, 0.05], T = [50, 100], functions = 200, sup=0.15):
    r = lambda x: kernel(0, x)

    Eps = [0.01, 0.05]#, 0.08, 0.15] # epsilon
    T = [50, 100]#, 5000, 10000] # time intervals 


    fig, axs = plt.subplots(len(Eps), len(T), figsize=(15, 15))

    for i, e in enumerate(Eps):
        for j, t in enumerate(T): 
            # Y -- массив из 200 функций, zeros -- массив нулей, E_zeros -- среднее число нулей
                                                   #mean, kernel, points, functions, left, right, iterations
            zeros, E_zeros, Y = post_intersections(zero, kernel, 40, functions, 0, t, t//4, ls = False, y = True) 
            overcrowding = np.zeros(Y.shape[1])
            Q = np.fft.fft(overcrowding) # будущий under_density
            Q = np.array([np.real(q) for q in Q])
            
            size = Q.shape[0]
            N = [i/size for i in range(-size//2, size//2)]
            count = 0

            for z in range(len(zeros)):
                if E_zeros - zeros[z] < -e*t:
                    Q0 =  np.fft.fft(Y[z])
                    Q0 = np.array([np.abs(q)**2 for q in Q0])
                    Q0 /= size
                    Q += Q0
                    count += 1
                
            Q /= count + 1e-30
            Q *= t/size/np.sqrt(2*np.pi)
            Q = np.hstack((Q[size//2:], Q[:size//2]))
            under_density = Q

            X = np.linspace(-t/2, t/2, size+1)[:-1]
            k = [r(x) for x in X]                  # kernel time points
            true_density = np.fft.fft(k)
            true_density = [abs(np.real(x)) for x in true_density]
            true_density = np.hstack((true_density[size//2:], true_density[:size//2]))
            true_density *= t/size/np.sqrt(2*np.pi)

            axs[i, j].plot(N, under_density)
            axs[i, j].plot(N, true_density)
            axs[i, j].set_xlim(-sup, sup)
            axs[i, j].set_title(f'Eps = {e},T = {t}')
```
